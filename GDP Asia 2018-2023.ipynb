```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ganti path file CSV sesuai dengan lokasinya
file_path = r"C:\Users\Kichiro\OneDrive\Desktop\Path Learning ML\Python for Data Science\Latihan data science\Dataset\ADO April 2023 - Fixing.csv"

# Membaca file CSV ke dalam DataFrame
dataset = pd.read_csv(file_path, encoding='ISO-8859-1,', delimiter = ";")
```

```python
# Ubah nama kolom
dataset.columns = ['RegionalMember', 'Year', 'GDO Growth', 'Unit of Measurement', 'Subregion', 'Country Code']
```

```python
dataset.head
```

```python
#memilih hanya negara, GDO pada tahun 2022

# Filter data untuk tahun 2022
data_2022 = dataset[dataset['Year'] == '2022']

# Mengelompokkan data berdasarkan negara dan menghitung rata-rata pertumbuhan GDP
gdo_growth_2022 = data_2022.groupby('RegionalMember')['GDO Growth'].mean().reset_index()
```

```python
# Mengurutkan data berdasarkan pertumbuhan GDP
gdo_growth_2022_sorted = gdo_growth_2022.sort_values(by='GDO Growth', ascending=False)
```

10 GDO Negara Tertinggi di dataset

```python
# Menampilkan 10 negara teratas dengan pertumbuhan GDP tertinggi
print(gdo_growth_2022_sorted.head(10))
```

```python
#mengurutkan

# Menambahkan kolom peringkat
gdo_growth_2022_sorted['Rank'] = range(1, len(gdo_growth_2022_sorted) + 1)

# Menampilkan 10 negara teratas dengan pertumbuhan GDP tertinggi
top_10_countries = gdo_growth_2022_sorted.head(10)

# Menampilkan hasil dengan peringkat yang sudah diurutkan
print(top_10_countries)
```

```python
# Plot diagram batang
plt.figure(figsize=(10, 6))
plt.barh(top_10_countries['RegionalMember'], top_10_countries['GDO Growth'], color='skyblue')
plt.xlabel('Pertumbuhan GDP (%)')
plt.ylabel('Negara')
plt.title('Top 10 Pertumbuhan GDO Teratas pada Tahun 2022')
plt.gca().invert_yaxis()  # Agar negara dengan pertumbuhan tertinggi berada di atas

# Menampilkan diagram
plt.show()
```

```python
# Data negara-negara peringkat 11-20 dengan pertumbuhan GDP tertinggi pada tahun 2022
rank_11_to_20 = gdo_growth_2022_sorted.iloc[10:20]

# Menampilkan DataFrame
print(rank_11_to_20)
```

```python
# Data negara-negara peringkat 11-20 dengan pertumbuhan GDP tertinggi pada tahun 2022
rank_11_to_20 = gdo_growth_2022_sorted.iloc[10:20]

# Plot diagram batang
plt.figure(figsize=(10, 6))
plt.barh(rank_11_to_20['RegionalMember'], rank_11_to_20['GDO Growth'], color='skyblue')
plt.xlabel('Pertumbuhan GDP (%)')
plt.ylabel('Negara')
plt.title('Peringkat 11-20 Pertumbuhan GDP pada Tahun 2022')
plt.gca().invert_yaxis()  # Agar negara dengan pertumbuhan tertinggi berada di atas

# Menampilkan diagram
plt.show()
```

GDO Indonesia

```python
# Filter data hanya untuk Indonesia
gdp_indonesia = dataset[dataset['RegionalMember'] == 'Indonesia']
```

```python
import matplotlib.pyplot as plt

# Plot diagram garis
plt.figure(figsize=(10, 6))
plt.plot(gdp_indonesia['Year'], gdp_indonesia['GDO Growth'], marker='o', color='blue', label='Pertumbuhan GDP')
plt.xlabel('Tahun')
plt.ylabel('Pertumbuhan GDP (%)')
plt.title('Perkembangan GDP Indonesia (2018-2024)')
plt.grid(True)
plt.legend()
plt.show()
```

Kesimpulan :

1. GDP Indonesia dari 2018 - 2024 (perkiraan) mengalami kenaikan dan penurunan, 
- terlihat pada tahun 2018 -2019 mengalami penurunan sebesar 0.2 point (dari 5.2 menjadi 5.0) 
- kemudian pada tahun 2019-2020 terlihat penurunan tajam sebesar 7.1 point  (5.0 menjadi -2.1)
- kemudian pada tahun 2020 - 2021 terlihat kenaikan dari titik -2.1 menjadi 3.7
- pada 2021 - 2022 mengalami kenaikan juga sebesar 1.6 point dari 3.7 sampai 5.3
- 2022 - 2023 mengalami penurunan yang landai yaitu sebesar 0.5 point yaitu dari 5.3 menjadi 4.8. 
- untuk prediksi 2023 - 2024 akan terjadi kenaikan sebesar 0.2 point dari 4.8 menjadi 5.0

Perbandingan 3 Negara ASEAN (Indonesia, Malaysia, Singapura) dan China

```python
# Filter data hanya untuk negara-negara yang ingin dibandingkan
countries_to_compare = ['Indonesia', 'Malaysia', 'Singapore', "People's Republic of China"]
gdp_comparison = dataset[dataset['RegionalMember'].isin(countries_to_compare)]
```

```python
# Plot diagram garis
plt.figure(figsize=(10, 6))

for country in countries_to_compare:
    country_data = gdp_comparison[gdp_comparison['RegionalMember'] == country]
    plt.plot(country_data['Year'], country_data['GDO Growth'], marker='o', label=country)

plt.xlabel('Tahun')
plt.ylabel('Pertumbuhan GDP (%)')
plt.title('Perkembangan GDP Negara-negara (2018-2024)')
plt.grid(True)
plt.legend()
plt.show()
```
